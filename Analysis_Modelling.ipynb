```python
X = data[["Years", "Job Rate"]]
Y = data["Annual Salary"]
```

```python
# we set a procentage for the training data and one for the testing data that the model will use for learning
from sklearn.model_selection import train_test_split # type: ignore
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# since x and y have the same size, we do not need to perform any sizing 
```

```python
# we are going to train a linear regression model
from sklearn.linear_model import LinearRegression # type: ignore
```

```python
lr = LinearRegression()
```

```python
# we train the model on the X_train and Y_train data
lr.fit(X_train, Y_train)
```

```python
# here we make predictions now using X_test data
predslr = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error # type: ignore
```

```python
mean_absolute_error(predslr, Y_test)
```
